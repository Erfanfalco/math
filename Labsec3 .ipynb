##linspace: 
در واقع تابعی است که فاصله بین دو نتقطه ابتدایی و انتهایی (که توسط کدنویس) تعریف میشود را به فواصل مساوی تقسیم می کنه. تعداد این نقاط به صورت دیفالت 50 تاست اما میتواند کمتر یا بیشتر هم باشد.

```python
x=np.linspace(-1, 3, 1000)

def f(x):
    return x**2 +1 -np.sin(x)
    
def g(x):
    return x**3+np.sqrt(x)
```

```python
plt.plot(x,f(x))
plt.axhline(color='red')
plt.fill_between(x, f(x), where=[(x>0) and (x<2) for x in x], color = 'pink')
```

```python
plt.plot(x,g(x))
plt.axhline(color='purple')
plt.fill_between(x, g(x), where=[(x>0) and (x<2) for x in x], color='red') 
```

```python
plt.plot(x,g(x))
plt.axhline(color='black')
plt.fill_between(x, g(x), where=[(x>0) and (x<2) for x in x], color='green', alpha=0.7)
```

```python
x=np.linspace(0,2,100)
plt.plot(x,g(x), marker='*', color='pink')
plt.axhline(color='black')
plt.fill_between(x, g(x), where=[(x>0) and (x<2) for x in x], color='red') 
```

```python
def h(x):
    return x**2 -2*x-3
def k(x):
    return 2*x-3
```

##plt.rcParams['figure.figsize'] = [a,b]:  دستور تغییر سایز تصویر خروجی کد

```python
x=np.linspace(0, 6, 1000)
plt.rcParams['figure.figsize'] = [10, 10]
plt.plot(x, h(x))
plt.plot(x, k(x))

x = sy.Symbol('x')
sy.solve(h(x) - k(x), x)
#plt.fill_between(x, h(x), k(x), where=[(x>0) and (x<2) for x in x], color='green', alpha=0.5)
```

سطح محصور بیم دو تابع $h(x)= x^2-2x-3$ و $k(x)= 2x-3$ را بیابید و در نمودار نشان دهید.

```python
x=np.linspace(0, 6, 1000)
plt.rcParams['figure.figsize'] = [10, 4]
plt.plot(x,h(x))
plt.plot(x,k(x))
plt.xlabel('$x^2$')
plt.fill_between(x, h(x), k(x), where=[(x>0) and (x<4) for x in x], color='green', alpha=0.3)
```
